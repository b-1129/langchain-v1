### stream and batch

```python
for chunk in model.stream("write me a benefits of eating eggs daily"):
    print(chunk.text, end="|", flush=True)
```

```python
responses = model.batch([
    "hey, what's going on?",
    "give me 5 businnes idea to earn 100 CR in today's world from zero",
    "how to be lean and fit forever?"
])
for response in responses:
    print(response)
```

```python
responses = model.batch([
    "hey, what's going on?",
    "give me 5 businnes idea to earn 100 CR in today's world from zero",
    "how to be lean and fit forever?"
],
config={"max_concurrency": 5})
for response in responses:
    print(response)
```
